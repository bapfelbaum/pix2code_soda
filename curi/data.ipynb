```python
# Collect the accuracies of all splits
import os
import json

SPLITS = [
    "color_boolean",
    "color_count",
    "color_location",
    "color_material",
    "color",
    "comp",
    "iid",
    "length_threshold_10",
    "shape",
]

accuracies = {}
for split in SPLITS:
    accuracies[split] = [0, 0, 0]

for split in SPLITS:
    folder = f"workspace/runs/{split}"
    for models in os.listdir(folder):
        if "alpha=1.0" in models:
            continue

        model_path = os.path.join(folder, models)

        for seed in [0, 1, 2]:
            metrics = model_path + f"/{seed}/test_metrics.txt"
            try:
                with open(metrics, "r") as file:
                    # Iterate through each line in the file
                    for line in file:
                        if "-1" in line and "modelmetrics/acc" in line:
                            _, _, acc = line.strip().split(",")
                            if float(acc) > accuracies[split][seed]:
                                accuracies[split][seed] = float(acc)
                    print(split, seed, acc)
            except FileNotFoundError:
                pass
```

```python
# get mean and std of accuracies

import numpy as np

for split in SPLITS:
    acc = accuracies[split]
    acc = np.array(acc) * 100
    acc = acc[:2]
    mean = np.mean(acc)
    mean = round(mean, 2)
    std = np.std(acc)
    std = round(std, 2)
    print(split)
    print(f"{mean} $\pm$ {std}")
```

```python
# Collect the accuracies of all splits
import os
import json
import numpy as np


SPLITS = [
    "iid",
]

accuracies = {}
for split in SPLITS:
    accuracies[split] = {
        "concat": [0, 0, 0],
        "gap": [0, 0, 0],
        "rel_net": [0, 0, 0],
        "trnsf": [0, 0, 0],
    }

for split in SPLITS:
    folder = f"workspace/runs_kand/{split}"
    for models in os.listdir(folder):
        if "pooling=concat" in models:
            pooling = "concat"
        elif "pooling=gap" in models:
            pooling = "gap"
        elif "pooling=rel_net" in models:
            pooling = "rel_net"
        elif "pooling=trnsf" in models:
            pooling = "trnsf"

        if "json" in models:
            continue

        model_path = os.path.join(folder, models)

        for seed in [0, 1, 2]:
            # metrics = model_path + f"/{seed}/test_metrics_all_metal_one_gray_8.txt"
            metrics = model_path + f"/{seed}/metrics.txt"
            try:
                with open(metrics, "r") as file:
                    # Iterate through each line in the file
                    for line in file:
                        if "999" in line and "modelmetrics/acc" in line:
                            _, _, acc = line.strip().split(",")
                    if float(acc) > accuracies[split][pooling][seed]:
                        accuracies[split][pooling][seed] = float(acc)
                    print(split, pooling, seed, acc)
            except FileNotFoundError:
                pass


for split in SPLITS:
    acc = accuracies[split]
    for pooling in ["concat", "gap", "rel_net", "trnsf"]:
        acc = accuracies[split][pooling]
        acc = np.array(acc) * 100
        mean = np.mean(acc)
        mean = round(mean, 2)
        std = np.std(acc)
        std = round(std, 2)
        print(split + " " + pooling)
        print(f"{mean} $\pm$ {std}")
```
