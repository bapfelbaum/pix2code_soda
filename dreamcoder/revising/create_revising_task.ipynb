```python
import os
import json

# Shift other indices so that each is unique
folder_path = "../data/curi/confounded/train/query/"
# get all files in folder
files = os.listdir(folder_path)

sizes = [0, 1]
colors = [2, 3, 4, 5, 6, 7, 8, 9]
shapes = [10, 11, 12]
materials = [13, 14]
for file in files:
    # load file
    with open(folder_path + file, "r") as f:
        task = json.load(f)
    for t in task:
        print(t)
        input = t["input"]
        for i in input:
            i[5] = colors[i[5]]
            i[6] = shapes[i[6]]
            i[7] = materials[i[7]]
        t["input"] = input

    # save to file
    with open(folder_path + file, "w") as f:
        json.dump(task, f)
```
